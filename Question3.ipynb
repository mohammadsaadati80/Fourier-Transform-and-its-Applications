### A

```python
signal = np.loadtxt("Signal.txt")
time = np.linspace(0 , 2 , 2000)

def sig(i):
    r = signal*np.exp(-2*np.pi*i*1j*time)
    return r.sum()/r.size

def fre(x, n):
    return np.array([2*np.exp(1j*2*i*np.pi*x)*sig(i) for i in range(1,n+1)]).sum()

y2 = np.array([fre(t,5).real for t in time])
plt.plot(time, signal , label = "main")
plt.plot(time, y2 , label = "recovered")
plt.legend()
plt.show()
```
