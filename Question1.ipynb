## Part 1 (Sine/Complex Sine Waves)

### A

```python
import numpy as np
import math
from matplotlib import pyplot as plt                 

sample_rate = 500
start_time = 0.
end_time = 2.
t_seq = np.arange(start_time ,end_time , 1./sample_rate)
frequency = 3
amplitude = 2
phase = np.pi/3

sinewave = amplitude*np.sin(2*np.pi*frequency*t_seq+phase)
plt.plot(t_seq,sinewave)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Sine Wave")
plt.show
```

### B

```python
complex_sine_wave = amplitude * np.exp( 1j* (2*np.pi * frequency * t_seq));
plt.plot(t_seq,np.real(complex_sine_wave),label='real')
plt.plot(t_seq,np.imag(complex_sine_wave),label='imag')
plt.xlabel('Time(s)'), plt.ylabel('Amplitude')
plt.title('Complex sine wave ')
plt.legend()
```

### C

```python
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure(figsize=(7,7))
ax = fig.gca(projection ='3d')
ax.plot(t_seq, complex_sine_wave.real, complex_sine_wave.imag)
ax.set_xlabel('Time(s)'), ax.set_ylabel('Real Part'), ax.set_zlabel('Imaginary Part')
plt.show()
```

## Part 2 (Dot/Complex Dot Product)

### A
<b>Dot product.</b> The dot product is a single number that reflects the relationship between two vectors (a vector is a list of numbers). The dot product is the computational backbone of many algorithms and mathematical operations, including the Fourier transform, convolution, and correlation. To compute the dot product between two vectors, simply point-wise multiply the corresponding elements and then sum over all elements. The larger the dot product, the more similar the two vectors are to each other (after accounting for the scaling of each vector, which might be arbitrary).

```python
time = np.arange(-1., 1., 1/1000.)
theta = 0
signal = np.sin(2*np.pi*5*time + theta) * np.exp( (-time**2) / .1)

plt.plot(time,signal)
plt.ylabel('Amplitude (AU)')
plt.xlabel("Time(s)")
plt.title("f(t)")
plt.show
```

#### 2

```python
fval = np.arange(2.,10.,.5);
dps = np.zeros(len(fval));
for f in range(1,len(dps)):
    sineWave = np.sin( 2*np.pi*fval[f]*time)
    dps[f] = np.dot(sineWave,signal)/len(time)

plt.stem(fval,dps)
plt.xlabel('Sine wave frequency'), plt.ylabel('Dot product')
plt.title('Sine waves dot products')
plt.show()
```

The maximum product in plot is for the wave with frequency 5 because the frequency is equal to f(t) frequency

```python
theta = np.pi/4;
signal = np.multiply(np.sin(2*np.pi*5*time + theta) ,np.exp( (-time**2) / .1))
dps = np.dot(np.sin(2*np.pi*5*time) ,signal)/len(time)
dpc = np.dot(np.cos(2*np.pi*5*time) ,signal)/len(time)
dpComplex = np.complex(dpc,dps) 
mag = np.abs(dpComplex)
phs = np.angle(dpComplex)
plt.plot( dpc , dps ,'ro')
plt.xlabel('Cos axis')
plt.ylabel('Sin axis')
plt.axis('square')
plt.grid(True)
plt.axis([-.2,.2,-.2,.2])
plt.show()
plt.polar([phs,phs],[0,mag])
plt.show()
```

```python
import pylab as pl
from IPython import display
import time as ttime

phases = np.linspace(0,7*np.pi/2,num=100)
complex_sine_wave = np.exp(1j*2*np.pi*5*time)
real_sine_wave = np.sin(2*np.pi*5*time )

for phi in range(0,len(phases)):
    signal = np.multiply(np.sin(2*np.pi*5*time + phases[phi]) ,np.exp( (-time**2) / .1))
    cdp = np.sum(np.multiply(signal,complex_sine_wave))/len(time)
    rdp = sum(np.multiply(signal,real_sine_wave))/len(time)
    pl.cla()
    plt.subplot2grid((2,2), (0, 0), colspan=2)
    plt.plot(time,signal)
    plt.plot(time,real_sine_wave)
    plt.title('Signal and sine wave over time')
    plt.subplot2grid((2,2), (1, 0))
    plt.plot(np.real(cdp),np.imag(cdp),'ro')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.axis('square')
    plt.grid(True)
    plt.axis([-.2,.2,-.2,.2])
    plt.plot([0,np.real(cdp)],[0,np.imag(cdp)],'r')
    plt.subplot2grid((2,2), (1, 1))
    plt.plot(rdp,0,'ro')
    plt.xlabel('Real')
    plt.axis('square')
    plt.grid(True)
    plt.axis([-.2,.2,-.2,.2])  
    display.clear_output(wait=True)
    display.display(pl.gcf())
    ttime.sleep(.01)
```
